```python
# Análise de Vendas Mensais
df.head()
```

```python
# Análise de Lucro Total e Total de Vendas
df["custo_total"] = df["Custo_Produto"] * df["Quantidade_Vendida"] # Calcula o custo total multiplicando o custo do produto pela quantidade vendida
df["lucro"] = df["Valor_Total_Venda"] - df["custo_total"] # Calcula o lucro subtraindo o custo total do valor total de venda
df["resultado"] = df["lucro"].apply(lambda x: "Lucro" if x > 0 else "Prejuízo") # Classifica o resultado como "Lucro" ou "Prejuízo" com base no valor do lucro
df.head()
```

```python
# Análise de Lucro Total e Total de Vendas
lucro_total = df["lucro"].sum()
total_vendas = df["Valor_Total_Venda"].sum()
print(f"Lucro total: R$ {lucro_total:,.2f}")
print(f"Total de vendas: R$ {total_vendas:,.2f}")
```

```python
# Análise de Vendas Mensais
vendas_mensais = df.groupby(pd.Grouper(key="Data_Venda", freq="M"))["Valor_Total_Venda"].sum()
vendas_mensais = pd.DataFrame({
    "Data_Venda": vendas_mensais.index,
    "Valor_Total_Venda": vendas_mensais.values.round(2)
})

vendas_mensais
```

```python
# Cálculo da média móvel de 3 meses para as vendas mensais
# A média móvel de 3 meses é calculada usando a função rolling() do pandas, que cria uma janela deslizante de 3 meses sobre os dados de vendas mensais. A função mean() é aplicada para calcular a média dentro dessa janela, e o resultado é arredondado para 2 casas decimais usando round(2).
# A média móvel é uma técnica de suavização que ajuda a identificar tendências subjacentes nos dados, reduzindo o impacto de flutuações sazonais ou ruídos. No contexto das vendas mensais, a média móvel de 3 meses pode ajudar a destacar tendências de crescimento ou declínio nas vendas ao longo do tempo, proporcionando uma visão mais clara do desempenho geral da loja.
# O resultado é um DataFrame que contém as datas de venda e as respectivas médias móveis de vendas mensais, permitindo uma análise mais aprofundada das tendências de vendas ao longo do tempo.
vendas_mensais_ma = vendas_mensais.set_index("Data_Venda")["Valor_Total_Venda"].rolling(3).mean().round(2)
vendas_mensais_ma

vendas_mensais_ma = pd.DataFrame({
    "Data_Venda": vendas_mensais_ma.index,
    "Média_Móvel_3_Meses": vendas_mensais_ma.values
})

vendas_mensais_ma
```

```python
# Criar um gráfico combinando vendas mensais e média móvel
fig = px.bar(
    x=vendas_mensais['Data_Venda'],
    y=[vendas_mensais['Valor_Total_Venda'], vendas_mensais_ma['Média_Móvel_3_Meses']],
    title='Vendas Mensais vs Média Móvel Simples (3 meses)',
    labels={'x': 'Mês', 'value': 'Valor (R$)'},
    barmode='group',
    template='plotly_dark'
)

# Adicionar valores formatados como moeda nas barras
fig.data[0].text = 'R$ ' + (vendas_mensais['Valor_Total_Venda'].round(2).astype(str))
fig.data[0].textposition = 'inside'
fig.data[0].name = 'Vendas Mensais'

fig.data[1].text = 'R$ ' + (vendas_mensais_ma['Média_Móvel_3_Meses'].round(2).astype(str))
fig.data[1].textposition = 'inside'
fig.data[1].name = 'Média Móvel Simples(3 meses)'

fig.show()
```

```python
# Análise de Lucro por Produto
lucro_produto = (
    df.groupby("Nome_Produto")["lucro"]# Agrupa os dados por nome do produto e seleciona a coluna de lucro
    .sum() # Calcula a soma do lucro para cada produto
    .reset_index()   # Reseta o índice para transformar o resultado em um DataFrame
    .sort_values(by="lucro", ascending=False) # Ordena os produtos pelo lucro total em ordem decrescente
    
)

lucro_produto = pd.DataFrame({
    "Nome_Produto": lucro_produto["Nome_Produto"], # Seleciona a coluna de nome do produto
    "Lucro_Total (R$)":lucro_produto["lucro"].round(2) # Seleciona a coluna de lucro total e arredonda para 2 casas decimais
})

lucro_produto.head(10) # Exibe os 10 produtos com maior lucro total
```

```python
# Criar funil com valores formatados em moeda brasileira (R$) sem duplicação
top10 = lucro_produto.head(10).copy()
top10['Lucro_Formatado'] = top10['Lucro_Total (R$)'].apply(
    lambda v: f"R$ {v:,.2f}".replace(',', 'X').replace('.', ',').replace('X', '.')
)

fig_funil = px.funnel(
    top10,
    x='Lucro_Total (R$)',
    y='Nome_Produto',
    template='plotly_dark',
    title='TOP 10 - Produtos mais lucrativos',
    labels={'Lucro_Total (R$)': 'Lucro Total (R$)', 'Nome_Produto': 'Nome do Produto'} # Personalizar os rótulos dos eixos
)

# Exibir apenas o texto formatado dentro dos funis e remover o valor bruto do hover
fig_funil.data[0].text = top10['Lucro_Formatado'].values # Exibir apenas o texto formatado dentro dos funis
fig_funil.data[0].textposition = 'inside'# Exibir apenas o texto formatado dentro dos funis
fig_funil.update_traces(texttemplate='%{text}', hovertemplate='Produto=%{y}<extra></extra>') # Remove o valor bruto do hover

fig_funil.show()
```
